# Run Time Error in Python
### ZeroDivisionError, TypeError etc

```python

a:int = int(input("Enter first digit: \t"))
b:int = int(input("Enter second digit:\t"))

try :
   
    res = a/b
    print(res)
except(ZeroDivisionError):
    print("Can't divide by zero")
    print("please input correct digits")
    
```

```python
ind = ["mohsin","ahsan","faisal"]

try:
    a = ind[5]
    print(10/0)
except(ZeroDivisionError,IndexError):
    print("cant divide by zeror")
    print("Index cant found")
```

```python
ind = ["mohsin","ahsan","faisal"]
print("hellow first")
try:
    print(10/0)
    a = ind[5]
   
except(ZeroDivisionError):
    print("cant divide by zeror")
except IndexError:
    print("Index cant found")
except:
    print("something's wrong")
    
    
print("Hellow last")
```

# Custom error handling


```python
class StudentAgeError(Exception):
    pass

class StudentInfo():
    def __init__(self, name:str, roll_no:int, age:int) -> None:
        if age < 18 or age >60:
            raise StudentAgeError("You're not eligible")
        self.name =name
        self.roll_no=roll_no
        self.age=age
            

            
newStudent = StudentInfo("Mohsin","AIC4100",27)
print(newStudent.name, newStudent.roll_no, newStudent.age)
        
```

## Read pdf file

```python
from typing import List
import PyPDF2

def my_pdf(file_path:str)->List[str]:
    with open(file_path,"rb") as file:
        reader: PyPDF2.PdfReader = PyPDF2.PdfReader(file )
        text_content: List[str] = [page.extract_text() for page in reader.pages]
        return text_content

a:List[str] = my_pdf("./mypdf.pdf")
print(a)
```
